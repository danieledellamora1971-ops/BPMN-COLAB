````python
# -*- coding: utf-8 -*-
"""
BPMN GENERATOR - versione GOOGLE COLAB
======================================
Descrivi un processo in linguaggio naturale, un LLM (Ollama) lo converte in
logica, l'auto-layout calcola le coordinate e il motore BPMN esporta in
BPMN 2.0 XML, draw.io XML e SVG.

DIFFERENZE RISPETTO ALLA VERSIONE DESKTOP:
  - Niente Tkinter (Colab non ha display): l'interfaccia usa ipywidgets.
  - Ollama NON gira in Colab: devi puntare a un Ollama raggiungibile via URL
    (es. la tua macchina esposta con ngrok, oppure Ollama installato nella VM
    Colab e servito). Vedi la sezione CONFIG -> OLLAMA_URL.
  - I file si scaricano con google.colab.files.download() e si mostrano inline.
  - L'SVG viene renderizzato direttamente nel notebook (IPython.display.SVG).

COME USARLO IN COLAB (in celle separate):

  # Cella 1 - dipendenze
  !pip install requests ipywidgets -q

  # Cella 2 - carica questo file
  # (caricalo con il pannello File di Colab, oppure incolla tutto il codice)
  # poi:
  from bpmn_generator_colab import avvia
  avvia()

OPZIONE A - Ollama sulla TUA macchina esposto con ngrok:
  Sul tuo PC:
      ollama serve
      ngrok http 11434
  Copia l'URL https://xxxx.ngrok-free.app e mettilo nel campo "Ollama URL"
  della UI (aggiungendo /api/chat), oppure in Config.OLLAMA_URL qui sotto.
  NB: alcune versioni di Ollama+ngrok richiedono l'header host: se hai errori
  403, avvia Ollama con  OLLAMA_HOST=0.0.0.0 ollama serve.

OPZIONE B - Ollama DENTRO la VM Colab (piu' lento, senza GPU dedicata al LLM):
      !curl -fsSL https://ollama.com/install.sh | sh
      import subprocess, time
      subprocess.Popen(["ollama", "serve"])
      time.sleep(5)
      !ollama pull qwen2.5:3b
  In questo caso OLLAMA_URL resta http://localhost:11434/api/chat
"""

import json
import uuid
import xml.etree.ElementTree as ET
from dataclasses import dataclass, field
from typing import List, Optional, Dict

try:
    import requests
    HAS_REQUESTS = True
except ImportError:
    HAS_REQUESTS = False


# ═══════════════════════════════════════════════════════
# CONFIG
# ═══════════════════════════════════════════════════════
class Config:
    # In Colab, Ollama locale NON esiste di default. Imposta qui l'URL del tuo
    # Ollama (via ngrok) OPPURE usa l'OPZIONE B (Ollama nella VM Colab).
    OLLAMA_URL = "http://localhost:11434/api/chat"
    OLLAMA_MODEL = "gemma4:e4b"      # nome ESATTO come in `ollama list`
    OLLAMA_TIMEOUT = 600

    # Layout auto (identico alla versione desktop)
    NODE_W_TASK = 120
    NODE_H_TASK = 60
    NODE_GW = 50
    NODE_EV = 40
    COL_GAP = 60
    LANE_H = 160
    LANE_HEADER = 30
    POOL_X = 30
    POOL_Y = 30
    MARGIN_X = 40
    MARGIN_Y = 50


# ═══════════════════════════════════════════════════════
# MODELLO DATI  (invariato)
# ═══════════════════════════════════════════════════════
NS = "http://www.omg.org/spec/BPMN/20100524/MODEL"
DI = "http://www.omg.org/spec/BPMN/20100524/DI"
DC = "http://www.omg.org/spec/DD/20100524/DC"
DX = "http://www.omg.org/spec/DD/20100524/DI"


@dataclass
class BShape:
    id: str
    name: str
    kind: str
    x: float
    y: float
    w: float = 100
    h: float = 60


@dataclass
class BFlow:
    id: str = field(default_factory=lambda: "sf_" + uuid.uuid4().hex[:8])
    src: str = ""
    tgt: str = ""
    name: str = ""


@dataclass
class BLane:
    id: str
    name: str
    x: float
    y: float
    w: float
    h: float
    shapes: List[BShape] = field(default_factory=list)


@dataclass
class BPool:
    id: str
    name: str
    x: float
    y: float
    w: float
    h: float
    lanes: List[BLane] = field(default_factory=list)


@dataclass
class BProcess:
    id: str = field(default_factory=lambda: "proc_" + uuid.uuid4().hex[:8])
    name: str = "Processo BPMN"
    pools: List[BPool] = field(default_factory=list)
    flows: List[BFlow] = field(default_factory=list)

    def all_shapes(self) -> List[BShape]:
        return [s for p in self.pools for l in p.lanes for s in l.shapes]

    def shape_by_id(self, sid: str) -> Optional[BShape]:
        return next((s for s in self.all_shapes() if s.id == sid), None)


# ═══════════════════════════════════════════════════════
# 1) PROMPT + CLIENT OLLAMA  (invariato)
# ═══════════════════════════════════════════════════════
SYSTEM_PROMPT = """Sei un analista di processi esperto in notazione BPMN 2.0.
Ricevi la descrizione di un processo in linguaggio naturale e la converti in un
oggetto JSON che ne rappresenta la LOGICA. NON calcolare coordinate.

Regole:
- Individua le CORSIE (lanes): i ruoli/reparti/sistemi che eseguono le attivita'.
  Se non e' chiaro, usa una sola lane chiamata "Processo".
- Ogni nodo ha: "id" (breve, univoco, senza spazi), "name" (etichetta leggibile),
  "kind" fra: start, end, task, subprocess, gateway_xor (decisione esclusiva),
  gateway_and (parallelo), gateway_or (inclusivo), "lane" (nome della corsia).

REGOLA CRITICA SUI FINALI (non violarla mai):
- Ci deve essere almeno UNO start e almeno UNO end.
- OGNI ramo che conclude il processo DEVE terminare con un nodo "kind":"end".
- Se il testo dice "il processo termina", "l'ordine viene respinto", "il cliente
  conferma la ricezione" o simili, quello e' un nodo "end". NON dimenticarlo.
- Ogni percorso possibile (anche i rami di rifiuto/errore) deve arrivare a un "end".
  Prima di rispondere, controlla: ogni finale del racconto ha il suo nodo end?

REGOLE SUI FLUSSI:
- I "flows" collegano i nodi: {"src": id, "tgt": id, "name": etichetta}.
- L'etichetta sui flussi serve SOLO se escono da un gateway (es. "Si"/"No").
- Dopo un gateway_xor devono uscire almeno 2 flussi con etichette diverse.

COSTRUTTI AVANZATI (usali quando il testo li descrive):
- ATTIVITA' PARALLELE ("contemporaneamente", "nel frattempo", "insieme"):
  apri con un gateway_and, poi i rami paralleli, poi RICHIUDI con un secondo
  gateway_and prima di proseguire.
- CICLO / RILAVORAZIONE ("torna indietro", "ripete", "rilavorazione"):
  crea un flusso che dal gateway di controllo TORNA a un nodo precedente.

Rispondi SOLO con JSON valido, senza testo prima o dopo, senza backticks.
Schema:
{
  "process_name": "string",
  "lanes": ["Lane A", "Lane B"],
  "nodes": [
    {"id":"start1","name":"Inizio","kind":"start","lane":"Lane A"},
    {"id":"t1","name":"Fai X","kind":"task","lane":"Lane A"},
    {"id":"end1","name":"Fine","kind":"end","lane":"Lane A"}
  ],
  "flows": [
    {"src":"start1","tgt":"t1","name":""},
    {"src":"t1","tgt":"end1","name":""}
  ]
}"""


def chiama_ollama(descrizione: str) -> Dict:
    """Chiama Ollama (endpoint /api/chat) e ritorna il dict JSON della logica."""
    if not HAS_REQUESTS:
        raise RuntimeError("Manca 'requests': !pip install requests")

    payload = {
        "model": Config.OLLAMA_MODEL,
        "messages": [
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user",
             "content": f"DESCRIZIONE PROCESSO:\n{descrizione}\n\nRispondi solo con JSON."},
        ],
        "stream": False,
        "format": "json",
        "options": {"temperature": 0.1, "num_ctx": 4096},
    }
    r = requests.post(Config.OLLAMA_URL, json=payload, timeout=Config.OLLAMA_TIMEOUT)
    r.raise_for_status()
    data = r.json()
    raw = (data.get("message", {}) or {}).get("content", "").strip()
    if not raw:
        raw = data.get("response", "").strip()
    raw = raw.replace("```json", "").replace("```", "").strip()
    return json.loads(raw)


# ═══════════════════════════════════════════════════════
# 2) VALIDAZIONE  (invariata)
# ═══════════════════════════════════════════════════════
_KINDS = {"start", "end", "task", "subprocess",
          "gateway_xor", "gateway_and", "gateway_or"}


def auto_fix(d: Dict) -> List[str]:
    fix = []
    nodes = d.get("nodes", [])
    flows = d.get("flows", [])
    if not nodes:
        return fix

    # --- FIX: flussi che puntano a un "kind" invece che a un id nodo ---
    # Gemma/altri modelli a volte usano "gateway_and" come id nel flusso invece
    # dell'id reale del gateway (es. "gw1"). Se un flusso cita un id inesistente
    # che pero' coincide con il kind di un nodo, lo rimappiamo all'id di QUEL nodo.
    ids_reali = {n.get("id") for n in nodes}
    kind_to_ids: Dict[str, List[str]] = {}
    for n in nodes:
        kind_to_ids.setdefault(n.get("kind"), []).append(n.get("id"))

    def _rimappa(valore, contesto):
        # valore inesistente ma uguale a un kind con UN solo nodo di quel tipo?
        if valore in ids_reali:
            return valore
        cand = kind_to_ids.get(valore, [])
        if len(cand) == 1:
            fix.append(f"flusso {contesto} '{valore}' -> rimappato su id '{cand[0]}'")
            return cand[0]
        return valore  # ambiguo o non risolvibile: lo lascia, la validazione segnalera'

    for f in flows:
        if f.get("src") not in ids_reali:
            f["src"] = _rimappa(f.get("src"), "src")
        if f.get("tgt") not in ids_reali:
            f["tgt"] = _rimappa(f.get("tgt"), "tgt")

    con_uscita = {f.get("src") for f in flows}
    if not any(n.get("kind") == "end" for n in nodes):
        candidati = [n for n in nodes
                     if n.get("id") not in con_uscita
                     and n.get("kind") not in ("start",)]
        for n in candidati:
            n["kind"] = "end"
            fix.append(f"nodo '{n.get('id')}' senza uscite -> marcato come end")
        if not candidati:
            nuovo_id = "end_auto"
            lane = nodes[-1].get("lane")
            nodes.append({"id": nuovo_id, "name": "Fine",
                          "kind": "end", "lane": lane})
            ultimo = nodes[-2]["id"]
            flows.append({"src": ultimo, "tgt": nuovo_id, "name": ""})
            fix.append("aggiunto nodo end automatico in coda")
    return fix


def valida_logica(d: Dict) -> List[str]:
    errori = []
    nodes = d.get("nodes", [])
    flows = d.get("flows", [])
    if not nodes:
        errori.append("Nessun nodo generato.")
        return errori

    ids = [n.get("id") for n in nodes]
    if len(ids) != len(set(ids)):
        errori.append("ID nodi duplicati.")
    for n in nodes:
        if n.get("kind") not in _KINDS:
            errori.append(f"Nodo '{n.get('id')}': kind non valido '{n.get('kind')}'.")
    if not any(n.get("kind") == "start" for n in nodes):
        errori.append("Manca un nodo start.")
    if not any(n.get("kind") == "end" for n in nodes):
        errori.append("Manca un nodo end.")

    idset = set(ids)
    for f in flows:
        if f.get("src") not in idset:
            errori.append(f"Flusso con src inesistente: {f.get('src')}")
        if f.get("tgt") not in idset:
            errori.append(f"Flusso con tgt inesistente: {f.get('tgt')}")
    return errori


# ═══════════════════════════════════════════════════════
# 3) AUTO-LAYOUT  (invariato)
# ═══════════════════════════════════════════════════════
def _rank_nodi(nodes: List[Dict], flows: List[Dict]) -> Dict[str, int]:
    succ: Dict[str, List[str]] = {n["id"]: [] for n in nodes}
    for f in flows:
        if f["src"] in succ:
            succ[f["src"]].append(f["tgt"])

    rank: Dict[str, int] = {}
    start_ids = [n["id"] for n in nodes if n["kind"] == "start"] or [nodes[0]["id"]]
    frontiera = [(sid, 0) for sid in start_ids]
    visti = set()
    while frontiera:
        nid, r = frontiera.pop(0)
        if nid in visti and rank.get(nid, 0) >= r:
            continue
        visti.add(nid)
        rank[nid] = max(rank.get(nid, 0), r)
        for t in succ.get(nid, []):
            frontiera.append((t, r + 1))
    for n in nodes:
        rank.setdefault(n["id"], max(rank.values(), default=0) + 1)
    return rank


def costruisci_bprocess(d: Dict) -> BProcess:
    proc = BProcess(name=d.get("process_name", "Processo BPMN"))

    lanes_nomi = d.get("lanes") or ["Processo"]
    nodes = d["nodes"]
    flows = d["flows"]

    lane_index = {nome: i for i, nome in enumerate(lanes_nomi)}
    for n in nodes:
        ln = n.get("lane") or "Processo"
        if ln not in lane_index:
            lane_index[ln] = len(lane_index)

    rank = _rank_nodi(nodes, flows)
    max_rank = max(rank.values(), default=0)

    col_w = Config.NODE_W_TASK + Config.COL_GAP
    pool_w = Config.LANE_HEADER + Config.MARGIN_X + (max_rank + 1) * col_w + Config.MARGIN_X
    pool_h = len(lane_index) * Config.LANE_H

    pool = BPool(id="pool1", name=proc.name,
                 x=Config.POOL_X, y=Config.POOL_Y, w=pool_w, h=pool_h)

    lane_obj: Dict[str, BLane] = {}
    for nome, idx in sorted(lane_index.items(), key=lambda kv: kv[1]):
        ly = Config.POOL_Y + idx * Config.LANE_H
        lane = BLane(id=f"lane_{idx}", name=nome,
                     x=Config.POOL_X + Config.LANE_HEADER, y=ly,
                     w=pool_w - Config.LANE_HEADER, h=Config.LANE_H)
        lane_obj[nome] = lane

    id_map: Dict[str, BShape] = {}
    occ: Dict[tuple, int] = {}
    for n in nodes:
        kind = n["kind"]
        ln = n.get("lane") or "Processo"
        lane = lane_obj[ln]
        r = rank[n["id"]]

        if kind in ("start", "end"):
            w = h = Config.NODE_EV
        elif kind.startswith("gateway"):
            w = h = Config.NODE_GW
        else:
            w, h = Config.NODE_W_TASK, Config.NODE_H_TASK

        x = lane.x + Config.MARGIN_X + r * col_w
        k = (ln, r)
        slot = occ.get(k, 0)
        occ[k] = slot + 1
        y = lane.y + Config.MARGIN_Y + slot * (Config.NODE_H_TASK + 10)

        sh = BShape(id=n["id"], name=n["name"].replace("\\n", "\n"),
                    kind=kind, x=x, y=y, w=w, h=h)
        lane.shapes.append(sh)
        id_map[n["id"]] = sh

    pool.lanes = list(lane_obj.values())
    proc.pools = [pool]
    proc.flows = [BFlow(src=f["src"], tgt=f["tgt"], name=f.get("name", ""))
                  for f in flows]
    return proc


# ═══════════════════════════════════════════════════════
# 4) EXPORTERS  (invariati)
# ═══════════════════════════════════════════════════════
KIND_TAG = {
    "start": "startEvent", "end": "endEvent",
    "inter": "intermediateCatchEvent",
    "task": "task", "subprocess": "subProcess",
    "gateway_xor": "exclusiveGateway", "gateway_and": "parallelGateway",
    "gateway_or": "inclusiveGateway",
}


def export_bpmn(proc: BProcess, path: str):
    root = ET.Element(f"{{{NS}}}definitions",
                      {"targetNamespace": "http://example.com/bpmn", "id": "Definitions_1"})
    collab = ET.SubElement(root, f"{{{NS}}}collaboration", {"id": "collab1"})
    for pool in proc.pools:
        ET.SubElement(collab, f"{{{NS}}}participant",
                      {"id": pool.id, "name": pool.name, "processRef": proc.id})
    pe = ET.SubElement(root, f"{{{NS}}}process",
                       {"id": proc.id, "name": proc.name, "isExecutable": "false"})
    ls = ET.SubElement(pe, f"{{{NS}}}laneSet", {"id": "ls1"})
    for pool in proc.pools:
        for lane in pool.lanes:
            le = ET.SubElement(ls, f"{{{NS}}}lane", {"id": lane.id, "name": lane.name})
            for sh in lane.shapes:
                ET.SubElement(le, f"{{{NS}}}flowNodeRef").text = sh.id
    for sh in proc.all_shapes():
        ET.SubElement(pe, f"{{{NS}}}{KIND_TAG.get(sh.kind, 'task')}",
                      {"id": sh.id, "name": sh.name})
    for f_ in proc.flows:
        attrs = {"id": f_.id, "sourceRef": f_.src, "targetRef": f_.tgt}
        if f_.name:
            attrs["name"] = f_.name
        ET.SubElement(pe, f"{{{NS}}}sequenceFlow", attrs)

    diag = ET.SubElement(root, f"{{{DI}}}BPMNDiagram", {"id": "diag1"})
    plane = ET.SubElement(diag, f"{{{DI}}}BPMNPlane",
                          {"id": "plane1", "bpmnElement": "collab1"})

    def add_shape_di(eid, x, y, w, h, horizontal=False):
        el = ET.SubElement(plane, f"{{{DI}}}BPMNShape",
                           {"id": eid + "_di", "bpmnElement": eid})
        if horizontal:
            el.set("isHorizontal", "true")
        ET.SubElement(el, f"{{{DC}}}Bounds",
                      {"x": str(x), "y": str(y), "width": str(w), "height": str(h)})

    def add_edge_di(fid, pts):
        el = ET.SubElement(plane, f"{{{DI}}}BPMNEdge",
                           {"id": fid + "_di", "bpmnElement": fid})
        for wx, wy in pts:
            ET.SubElement(el, f"{{{DX}}}waypoint", {"x": str(wx), "y": str(wy)})

    for pool in proc.pools:
        add_shape_di(pool.id, pool.x, pool.y, pool.w, pool.h, horizontal=True)
        for lane in pool.lanes:
            add_shape_di(lane.id, lane.x, lane.y, lane.w, lane.h, horizontal=True)
            for sh in lane.shapes:
                add_shape_di(sh.id, sh.x, sh.y, sh.w, sh.h)
    for f_ in proc.flows:
        src = proc.shape_by_id(f_.src)
        tgt = proc.shape_by_id(f_.tgt)
        if src and tgt:
            add_edge_di(f_.id, [(src.x + src.w / 2, src.y + src.h / 2),
                                (tgt.x + tgt.w / 2, tgt.y + tgt.h / 2)])
    ET.indent(root, space="  ")
    with open(path, "wb") as fh:
        ET.ElementTree(root).write(fh, encoding="utf-8", xml_declaration=True)


GW_LABEL = {"gateway_xor": "X", "gateway_and": "+", "gateway_or": "O"}
DRAWIO_STYLES = {
    "start": "ellipse;whiteSpace=wrap;html=1;aspect=fixed;fillColor=#d5e8d4;strokeColor=#82b366;fontSize=10;",
    "end": "ellipse;whiteSpace=wrap;html=1;aspect=fixed;fillColor=#f8cecc;strokeColor=#b85450;strokeWidth=3;fontSize=10;",
    "task": "rounded=1;whiteSpace=wrap;html=1;fillColor=#dae8fc;strokeColor=#6c8ebf;fontSize=10;",
    "subprocess": "rounded=1;whiteSpace=wrap;html=1;arcSize=3;fillColor=#d5e8d4;strokeColor=#82b366;fontSize=10;",
    "gateway_xor": "rhombus;whiteSpace=wrap;html=1;fillColor=#fff2cc;strokeColor=#d6b656;fontSize=14;fontStyle=1;",
    "gateway_and": "rhombus;whiteSpace=wrap;html=1;fillColor=#dae8fc;strokeColor=#6c8ebf;fontSize=14;fontStyle=1;",
    "gateway_or": "rhombus;whiteSpace=wrap;html=1;fillColor=#f8cecc;strokeColor=#b85450;fontSize=14;fontStyle=1;",
}


def export_drawio(proc: BProcess, path: str):
    root = ET.Element("mxGraphModel",
                      {"grid": "1", "gridSize": "10", "guides": "1", "page": "1",
                       "pageWidth": "1654", "pageHeight": "1169"})
    parent = ET.SubElement(root, "root")
    ET.SubElement(parent, "mxCell", {"id": "0"})
    ET.SubElement(parent, "mxCell", {"id": "1", "parent": "0"})
    cid = [100]

    def nid():
        cid[0] += 1
        return str(cid[0])

    id_map = {}
    for pool in proc.pools:
        pool_cid = nid()
        pc = ET.SubElement(parent, "mxCell",
                           {"id": pool_cid, "value": pool.name,
                            "style": "shape=pool;startSize=20;horizontal=1;fillColor=#f5f5f5;strokeColor=#666666;fontSize=13;fontStyle=1;",
                            "vertex": "1", "parent": "1"})
        ET.SubElement(pc, "mxGeometry",
                      {"x": str(pool.x), "y": str(pool.y),
                       "width": str(pool.w), "height": str(pool.h), "as": "geometry"})
        for lane in pool.lanes:
            lane_cid = nid()
            lc = ET.SubElement(parent, "mxCell",
                               {"id": lane_cid, "value": lane.name,
                                "style": "swimlane;startSize=30;horizontal=0;fillColor=#f9f9f9;strokeColor=#82b366;fontSize=10;fontStyle=1;",
                                "vertex": "1", "parent": pool_cid})
            ET.SubElement(lc, "mxGeometry",
                          {"x": "0", "y": str(lane.y - pool.y),
                           "width": str(lane.w), "height": str(lane.h), "as": "geometry"})
            for sh in lane.shapes:
                sh_cid = nid()
                id_map[sh.id] = sh_cid
                label = GW_LABEL.get(sh.kind, sh.name.replace("\n", "&#xa;"))
                sc = ET.SubElement(parent, "mxCell",
                                   {"id": sh_cid, "value": label,
                                    "style": DRAWIO_STYLES.get(sh.kind, DRAWIO_STYLES["task"]),
                                    "vertex": "1", "parent": lane_cid})
                ET.SubElement(sc, "mxGeometry",
                              {"x": str(sh.x - lane.x), "y": str(sh.y - lane.y),
                               "width": str(sh.w), "height": str(sh.h), "as": "geometry"})
    for f_ in proc.flows:
        sc = id_map.get(f_.src)
        tc = id_map.get(f_.tgt)
        if sc and tc:
            ec = ET.SubElement(parent, "mxCell",
                               {"id": nid(), "value": f_.name,
                                "style": "edgeStyle=orthogonalEdgeStyle;html=1;rounded=1;endArrow=block;",
                                "edge": "1", "source": sc, "target": tc, "parent": "1"})
            ET.SubElement(ec, "mxGeometry", {"relative": "1", "as": "geometry"})
    ET.indent(root, space="  ")
    with open(path, "wb") as fh:
        ET.ElementTree(root).write(fh, encoding="utf-8", xml_declaration=False)


# ═══════════════════════════════════════════════════════
# EXPORTER SVG  (invariato) + helper per render inline
# ═══════════════════════════════════════════════════════
SVG_COLORS = {
    "start":       ("#d5e8d4", "#82b366"),
    "end":         ("#f8cecc", "#b85450"),
    "task":        ("#dae8fc", "#6c8ebf"),
    "subprocess":  ("#d5e8d4", "#82b366"),
    "gateway_xor": ("#fff2cc", "#d6b656"),
    "gateway_and": ("#dae8fc", "#6c8ebf"),
    "gateway_or":  ("#f8cecc", "#b85450"),
}
SVG_GW_LABEL = {"gateway_xor": "X", "gateway_and": "+", "gateway_or": "O"}


def _esc(s: str) -> str:
    return (s.replace("&", "&amp;").replace("<", "&lt;")
             .replace(">", "&gt;").replace('"', "&quot;"))


def _wrap_lines(text: str, max_chars: int = 16):
    righe = []
    for blocco in text.split("\n"):
        parole = blocco.split()
        cur = ""
        for w in parole:
            if len(cur) + len(w) + 1 <= max_chars:
                cur = (cur + " " + w).strip()
            else:
                if cur:
                    righe.append(cur)
                cur = w
        if cur:
            righe.append(cur)
    return righe or [text]


def build_svg(proc: BProcess) -> str:
    """Come export_svg ma ritorna la stringa SVG (utile per il render inline)."""
    pool = proc.pools[0]
    W = pool.x + pool.w + 40
    H = pool.y + pool.h + 40
    out = [f'<svg xmlns="http://www.w3.org/2000/svg" width="{W}" height="{H}" '
           f'viewBox="0 0 {W} {H}" font-family="Segoe UI, Arial, sans-serif">']
    out.append('<defs><marker id="arrow" markerWidth="10" markerHeight="10" '
               'refX="9" refY="3" orient="auto" markerUnits="strokeWidth">'
               '<path d="M0,0 L9,3 L0,6 Z" fill="#555"/></marker></defs>')
    out.append(f'<rect x="0" y="0" width="{W}" height="{H}" fill="#ffffff"/>')

    out.append(f'<rect x="{pool.x}" y="{pool.y}" width="{pool.w}" height="{pool.h}" '
               f'fill="none" stroke="#666" stroke-width="2"/>')
    out.append(f'<text x="{pool.x + 14}" y="{pool.y + pool.h/2}" '
               f'transform="rotate(-90 {pool.x + 14} {pool.y + pool.h/2})" '
               f'text-anchor="middle" font-size="13" font-weight="bold" '
               f'fill="#333">{_esc(pool.name)}</text>')

    for lane in pool.lanes:
        out.append(f'<rect x="{lane.x}" y="{lane.y}" width="{lane.w}" '
                   f'height="{lane.h}" fill="#fafafa" stroke="#82b366" '
                   f'stroke-width="1"/>')
        out.append(f'<text x="{lane.x + 16}" y="{lane.y + lane.h/2}" '
                   f'transform="rotate(-90 {lane.x + 16} {lane.y + lane.h/2})" '
                   f'text-anchor="middle" font-size="11" font-weight="bold" '
                   f'fill="#4a4a4a">{_esc(lane.name)}</text>')

    for f_ in proc.flows:
        s = proc.shape_by_id(f_.src)
        t = proc.shape_by_id(f_.tgt)
        if not s or not t:
            continue
        x1, y1 = s.x + s.w / 2, s.y + s.h / 2
        x2, y2 = t.x + t.w / 2, t.y + t.h / 2
        out.append(f'<line x1="{x1}" y1="{y1}" x2="{x2}" y2="{y2}" '
                   f'stroke="#555" stroke-width="1.5" marker-end="url(#arrow)"/>')
        if f_.name:
            mx, my = (x1 + x2) / 2, (y1 + y2) / 2
            out.append(f'<rect x="{mx-12}" y="{my-9}" width="24" height="16" '
                       f'fill="#ffffff" opacity="0.9"/>')
            out.append(f'<text x="{mx}" y="{my+3}" text-anchor="middle" '
                       f'font-size="10" fill="#b85450" '
                       f'font-weight="bold">{_esc(f_.name)}</text>')

    for sh in proc.all_shapes():
        fill, stroke = SVG_COLORS.get(sh.kind, SVG_COLORS["task"])
        cx, cy = sh.x + sh.w / 2, sh.y + sh.h / 2
        if sh.kind in ("start", "end"):
            r = sh.w / 2
            sw = 3 if sh.kind == "end" else 1.5
            out.append(f'<circle cx="{cx}" cy="{cy}" r="{r}" fill="{fill}" '
                       f'stroke="{stroke}" stroke-width="{sw}"/>')
            label_lines = _wrap_lines(sh.name, 10)
            ly = cy + sh.h / 2 + 12
            for i, ln in enumerate(label_lines):
                out.append(f'<text x="{cx}" y="{ly + i*11}" text-anchor="middle" '
                           f'font-size="9" fill="#333">{_esc(ln)}</text>')
        elif sh.kind.startswith("gateway"):
            pts = f"{cx},{sh.y} {sh.x+sh.w},{cy} {cx},{sh.y+sh.h} {sh.x},{cy}"
            out.append(f'<polygon points="{pts}" fill="{fill}" stroke="{stroke}" '
                       f'stroke-width="1.5"/>')
            out.append(f'<text x="{cx}" y="{cy+5}" text-anchor="middle" '
                       f'font-size="16" font-weight="bold" '
                       f'fill="{stroke}">{SVG_GW_LABEL.get(sh.kind,"X")}</text>')
            for i, ln in enumerate(_wrap_lines(sh.name, 12)):
                out.append(f'<text x="{cx}" y="{sh.y+sh.h+12 + i*11}" '
                           f'text-anchor="middle" font-size="9" '
                           f'fill="#333">{_esc(ln)}</text>')
        else:
            out.append(f'<rect x="{sh.x}" y="{sh.y}" width="{sh.w}" height="{sh.h}" '
                       f'rx="8" ry="8" fill="{fill}" stroke="{stroke}" '
                       f'stroke-width="1.5"/>')
            lines = _wrap_lines(sh.name, 16)
            total = len(lines)
            start_y = cy - (total - 1) * 6 + 4
            for i, ln in enumerate(lines):
                out.append(f'<text x="{cx}" y="{start_y + i*12}" '
                           f'text-anchor="middle" font-size="10" '
                           f'fill="#1a1a1a">{_esc(ln)}</text>')

    out.append("</svg>")
    return "\n".join(out)


def export_svg(proc: BProcess, path: str):
    with open(path, "w", encoding="utf-8") as fh:
        fh.write(build_svg(proc))


# ═══════════════════════════════════════════════════════
# 5) INTERFACCIA COLAB  (ipywidgets al posto di Tkinter)
# ═══════════════════════════════════════════════════════
def avvia():
    """Costruisce e mostra la UI ipywidgets nel notebook Colab."""
    import ipywidgets as widgets
    from IPython.display import display, SVG, clear_output

    # stato condiviso tra le callback
    stato = {"proc": None, "logica": None}

    # --- widget ---
    titolo = widgets.HTML(
        "<h2 style='color:#c8a96e;margin:0'>BPMN GENERATOR "
        "<span style='color:#888;font-size:14px'>(Colab)</span></h2>"
        "<div style='color:#555'>descrivi un processo &rarr; diagramma BPMN</div>")

    url_in = widgets.Text(
        value=Config.OLLAMA_URL, description="Ollama URL:",
        layout=widgets.Layout(width="520px"),
        style={"description_width": "90px"})
    model_in = widgets.Text(
        value="gemma4:e4b", description="Modello:",
        layout=widgets.Layout(width="320px"),
        style={"description_width": "90px"})

    descr_in = widgets.Textarea(
        value=(
            "Il cliente invia un ordine all'ufficio commerciale. Il commerciale "
            "registra l'ordine e verifica l'affidabilita' creditizia del cliente. "
            "Se il credito non e' approvato, l'ordine viene respinto e il cliente "
            "riceve una notifica di rifiuto. Se il credito e' approvato, il "
            "magazzino verifica la disponibilita' dei prodotti.\n"
            "Se i prodotti sono disponibili, contemporaneamente il magazzino "
            "prepara la merce e l'amministrazione emette la fattura; quando "
            "entrambe le attivita' sono concluse, la logistica organizza la "
            "spedizione.\n"
            "Se i prodotti non sono disponibili, la produzione crea un ordine di "
            "lavoro e fabbrica i pezzi. Il controllo qualita' ispeziona i pezzi "
            "prodotti: se non superano il controllo, tornano in produzione per la "
            "rilavorazione; se lo superano, passano al magazzino che prepara la "
            "merce e si procede con fatturazione e spedizione.\n"
            "Una volta spedita la merce, la logistica traccia la consegna. Il "
            "processo termina quando il cliente conferma la ricezione."),
        description="Processo:",
        layout=widgets.Layout(width="720px", height="150px"),
        style={"description_width": "90px"})

    btn_gen = widgets.Button(description="Genera BPMN", button_style="warning",
                             icon="cogs")
    btn_bpmn = widgets.Button(description="Scarica .bpmn", disabled=True, icon="download")
    btn_drawio = widgets.Button(description="Scarica .drawio", disabled=True, icon="download")
    btn_svg = widgets.Button(description="Scarica .svg", disabled=True, icon="download")

    log_out = widgets.Output(layout=widgets.Layout(
        border="1px solid #444", height="200px", overflow="auto",
        padding="6px", background_color="#0d1a2b"))
    diag_out = widgets.Output()

    def log(msg):
        with log_out:
            print(msg)

    # --- callback generazione ---
    def on_gen(_):
        Config.OLLAMA_URL = url_in.value.strip() or Config.OLLAMA_URL
        Config.OLLAMA_MODEL = model_in.value.strip() or Config.OLLAMA_MODEL
        descr = descr_in.value.strip()
        if not descr:
            log("[!] Scrivi la descrizione del processo.")
            return
        btn_gen.disabled = True
        log(f"\n[..] Interrogo Ollama ({Config.OLLAMA_MODEL}) su {Config.OLLAMA_URL} ...")
        try:
            logica = chiama_ollama(descr)
            log("[OK] Logica ricevuta dall'LLM.")
            for c in auto_fix(logica):
                log(f"    [auto-fix] {c}")
            errori = valida_logica(logica)
            if errori:
                log("[!] Problemi nella logica generata:")
                for e in errori:
                    log("    - " + e)
                log("    Riprova riformulando la descrizione.")
                return
            stato["logica"] = logica
            stato["proc"] = costruisci_bprocess(logica)
            proc = stato["proc"]
            log(f"[OK] Processo: {len(proc.all_shapes())} nodi, "
                f"{len(proc.pools[0].lanes)} corsie, {len(proc.flows)} flussi.")
            btn_bpmn.disabled = False
            btn_drawio.disabled = False
            btn_svg.disabled = False
            # render inline dell'SVG
            with diag_out:
                clear_output(wait=True)
                display(SVG(build_svg(proc)))
        except requests.exceptions.ConnectionError:
            log("[ERRORE] Ollama non raggiungibile.")
            log("         Verifica l'URL (ngrok?) o avvia Ollama nella VM Colab.")
        except requests.exceptions.Timeout:
            log("[ERRORE] Timeout: il modello e' troppo lento o non risponde.")
        except json.JSONDecodeError as e:
            log(f"[ERRORE] L'LLM non ha prodotto JSON valido: {e}")
            log("         Riprova o usa un modello piu' capace (es. qwen2.5:7b).")
        except Exception as e:
            log(f"[ERRORE] {type(e).__name__}: {e}")
        finally:
            btn_gen.disabled = False

    # --- callback download ---
    def _download(fmt):
        from google.colab import files
        proc = stato["proc"]
        if not proc:
            return
        nome = "processo." + fmt
        if fmt == "bpmn":
            export_bpmn(proc, nome)
        elif fmt == "drawio":
            export_drawio(proc, nome)
        elif fmt == "svg":
            export_svg(proc, nome)
        log(f"[OK] Generato {nome} -> avvio download.")
        files.download(nome)

    btn_gen.on_click(on_gen)
    btn_bpmn.on_click(lambda _: _download("bpmn"))
    btn_drawio.on_click(lambda _: _download("drawio"))
    btn_svg.on_click(lambda _: _download("svg"))

    # --- layout ---
    display(widgets.VBox([
        titolo,
        widgets.HBox([url_in, model_in]),
        descr_in,
        widgets.HBox([btn_gen, btn_bpmn, btn_drawio, btn_svg]),
        widgets.HTML("<b style='color:#c8a96e'>Diagramma:</b>"),
        diag_out,
        widgets.HTML("<b style='color:#c8a96e'>Log:</b>"),
        log_out,
    ]))


# ═══════════════════════════════════════════════════════
# USO DA SCRIPT PURO (senza UI) - comodo per test rapidi in Colab
# ═══════════════════════════════════════════════════════
def genera_da_testo(descrizione: str, mostra=True):
    """
    Pipeline completa senza widget: testo -> logica -> BProcess -> SVG.
    Ritorna il BProcess. Se mostra=True, renderizza l'SVG nel notebook.
    Utile per chiamate programmatiche in una cella.
    """
    logica = chiama_ollama(descrizione)
    auto_fix(logica)
    errori = valida_logica(logica)
    if errori:
        raise ValueError("Logica non valida: " + "; ".join(errori))
    proc = costruisci_bprocess(logica)
    if mostra:
        try:
            from IPython.display import display, SVG
            display(SVG(build_svg(proc)))
        except ImportError:
            pass
    return proc


# ==== AVVIO AUTOMATICO ====
avvia()
````
